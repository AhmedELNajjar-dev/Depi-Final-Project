## Removing Non‑Informative Metadata Columns

The following columns were removed:
**`Flow ID`, `Source IP`, `Destination IP`, `Source Port`, `Protocol`, and `Timestamp`.**

**Reason for removal:**
- These fields are **identifiers or contextual metadata**, not behavioral features of the network flow.  
- Machine‑learning models trained on these values would simply memorize host IPs or specific timestamps, leading to **data leakage** and **overfitting**, rather than learning actual traffic patterns.  

- `Protocol` and `Source Port` can vary by environment and provide little generalization value, while `Timestamp` and `Flow ID` are unique per record and therefore non‑predictive.


```python

drop_cols = ['Flow ID', 'Source IP', 'Destination IP',
             'Source Port', 'Protocol', 'Timestamp']

Thursday.drop(columns=drop_cols, inplace=True)
print("Columns removed:", drop_cols)
print("New shape:", Thursday.shape)
```
